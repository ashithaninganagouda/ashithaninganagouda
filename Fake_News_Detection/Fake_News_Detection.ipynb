```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
import re
import string
```

```python
data_fake = pd.read_csv("Fake.csv")
data_true = pd.read_csv("True.csv")
```

```python
data_fake.head()
```

```python
data_true.head()
```

```python
data_fake["class"] = 0
data_true["class"] = 1
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing = data_fake.tail(10)
for i in range(23480, 23470, -1):
    data_fake.drop([i], axis=0, inplace=True)

data_true_manual_testing = data_true.tail(10)
for i in range(21416, 21406, -1):
    data_true.drop([i], axis=0, inplace=True)
```

```python
data_fake.shape, data_true.shape
```

```python
data_fake_manual_testing["class"] = 0
data_true_manual_testing["class"] = 1
```

```python
data_fake_manual_testing.head(10)
```

```python
data_true_manual_testing.head(10)
```

```python
data_merge = pd.concat([data_fake, data_true], axis = 0)
data_merge.head(10)
```

```python
data_merge.columns
```

```python
data = data_merge.drop(["title", "subject", "date"], axis = 1)
```

```python
data.isnull().sum()
```

```python
data = data.sample(frac = 1)
```

```python
data.head()
```

```python
data.reset_index(inplace = True)
data.drop(["index"], axis = 1, inplace = True)
```

```python
data.columns
```

```python
data.head()
```

```python
def wordopt(text):
    text = text.lower()
    text = re.sub("\[.*?\]", "", text)
    text = re.sub("\\W", " ", text)
    text = re.sub("https?://\S+|www.\.\S+", "", text)
    text = re.sub("<.*?>+", "", text)
    text = re.sub("[%s]" % re.escape(string.punctuation), "", text)
    text = re.sub("\n", "", text)
    text = re.sub("\w*\d\w*", "", text)
    return text
```

```python
data["text"] = data["text"].apply(wordopt)
```

```python
x = data["text"]
y = data["class"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorization = TfidfVectorizer()
xv_train = vectorization.fit_transform(x_train)
xv_test = vectorization.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(xv_train, y_train)
```

```python
pred_lr = lr.predict(xv_test)
```

```python
lr.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_lr))
```

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(xv_train, y_train)
```

```python
pred_dt = dt.predict(xv_test)
```

```python
dt.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_dt))
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state = 0)
gb.fit(xv_train, y_train)
```

```python
pred_gb = gb.predict(xv_test)
```

```python
gb.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_gb))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(random_state = 0)
rf.fit(xv_train, y_train)
```

```python
pred_rf = rf.predict(xv_test)
```

```python
rf.score(xv_test, y_test)
```

```python
print(classification_report(y_test, pred_rf))
```

```python
def output_label(n):
    if n==0:
        return "Fake News"
    else:
        return "True News"


def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_text = pd.DataFrame(testing_news)
    new_def_text["text"] = new_def_text["text"].apply(wordopt)
    new_x_text = new_def_text["text"]
    new_xv_test = vectorization.transform(new_x_text)
    pred_lr = lr.predict(new_xv_test)
    pred_dt = dt.predict(new_xv_test)
    pred_gb = gb.predict(new_xv_test)
    pred_rf = rf.predict(new_xv_test)
    return print("\n\nLR Prediction: {}\nDT Prediction: {}\nGB Prediction: {}\nRF Prediction: {} ".format(output_label(pred_lr[0]),
                                                                                                          output_label(pred_dt[0]),
                                                                                                          output_label(pred_gb[0]),
                                                                                                          output_label(pred_rf[0])))
    
```

```python
news = str(input())
manual_testing(news)
```
